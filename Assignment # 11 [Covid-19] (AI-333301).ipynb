4. Create Function that take dataset columns and replace whitespace 
with Underscore and update dataset automatically and the pass the 
dataset.

```python
def replace_whitespace_with_underscore(df):
    
    original_columns = list(df.columns)

    new_columns = {col: col.replace(" ", "_") for col in original_columns}

    df.rename(columns=new_columns, inplace=True)  

    return df

Data = pd.read_csv('covid_19_country_wise_latest.csv', index_col=0)
Data.index.name = 'index'

Data = replace_whitespace_with_underscore(Data)

Data.tail()
```

6. fill the nan values with the mean of the column 

```python
for column in Data.columns:
    if Data[column].isnull().any():  
        mean_value = Data[column].mean()  
        Data[column].fillna(mean_value, inplace=True) 


Data.head()
```

20. Identify countries where total cases are increasing but death 
rates remain low.

```python
data = {
    'Country': ['CountryA', 'CountryB', 'CountryC', 'CountryA', 'CountryB', 'CountryC'],
    'Date': ['2023-12-20', '2023-12-20', '2023-12-20', '2023-12-27', '2023-12-27', '2023-12-27'],
    'Total_Cases': [1000, 500, 200, 1200, 520, 210],
    'Deaths': [10, 5, 2, 12, 5, 2]
}
df = pd.DataFrame(data)

df['Case_Change'] = df.groupby('Country')['Total_Cases'].diff()
df['Death_Rate'] = (df['Deaths'] / df['Total_Cases']) * 100
df['Death_Rate_Change'] = df.groupby('Country')['Death_Rate'].diff()

increasing_cases_low_deaths = df[
    (df['Case_Change'] > df['Total_Cases'].shift() * 0.1) & 
    (df['Death_Rate'] < 2)
]

countries = increasing_cases_low_deaths['Country'].unique()

print("Countries with increasing cases and low death rates:")
print(countries)
```
